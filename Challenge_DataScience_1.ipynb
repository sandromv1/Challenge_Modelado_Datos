## 2.1 Ventas por categoría

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Agregamos la columna 'Tienda' a cada DataFrame
tienda['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'

# Unimos todas las tiendas en un solo DataFrame
todas_tiendas = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)

# Aseguramos que el precio sea numérico
todas_tiendas['Precio'] = pd.to_numeric(todas_tiendas['Precio'], errors='coerce')

# Calculamos ingresos por categoría y tienda
ingresos_por_categoria_tienda = (
    todas_tiendas
    .groupby(['Tienda', 'Categoría del Producto'])['Precio']
    .sum()
    .reset_index()
    .sort_values(['Tienda', 'Precio'], ascending=[True, False])
)

# Mostrar tabla por cada tienda
for tienda_nombre, df_tienda in ingresos_por_categoria_tienda.groupby('Tienda'):
    print(f"\n📊 {tienda_nombre}")
    print(df_tienda.to_string(index=False))
```

# 3. Calificación promedio de la tienda


```python
# Diccionario con los DataFrames
tiendas = {
    'tienda': tienda,
    'tienda2': tienda2,
    'tienda3': tienda3,
    'tienda4': tienda4
}

# Iterar sobre cada tienda
for nombre, df in tiendas.items():
    cali_prom = df['Calificación'].value_counts()
    print(f"Calificación promedio de {nombre}: {cali_prom}")
```

```python
for nombre, df in tiendas.items():
    cali_counts = df['Calificación'].value_counts().sort_index()
    plt.figure(figsize=(5, 5))
    plt.pie(cali_counts, labels=cali_counts.index, autopct='%1.1f%%', startangle=90)
    plt.title(f"Distribución porcentual de calificaciones - {nombre.upper()}")
    plt.axis('equal')  # Círculo perfecto
    plt.show()
```

```python
# Iterar sobre cada tienda
for nombre, df in tiendas.items():
    cali_prom = df['Calificación'].value_counts()
    calificacion_promedio = (cali_prom * cali_prom.index).sum() / cali_prom.sum()
    print(f"Calificación promedio de {nombre}: {calificacion_promedio:.2f}")
```

# 4. Productos más y menos vendidos

```python
tiendas = {
    'tienda': tienda,
    'tienda2': tienda2,
    'tienda3': tienda3,
    'tienda4': tienda4
}

for nombre, df in tiendas.items():
    print(f"\n📦 Resultados para: {nombre.upper()}")

    ventas = df['Producto'].value_counts()

    print("🟢 Más vendidos:")
    print(ventas.head(3))

    print("🔴 Menos vendidos:")
    print(ventas.tail(3))
```

```python
import matplotlib.pyplot as plt

tiendas = {
    'tienda': tienda,
    'tienda2': tienda2,
    'tienda3': tienda3,
    'tienda4': tienda4
}

for nombre, df in tiendas.items():
    ventas = df['Producto'].value_counts().sort_values(ascending=True).tail(5)  # Orden ascendente para barras horizontales

    plt.figure(figsize=(10,4))
    ventas.plot(kind='barh', color='skyblue', edgecolor='black')

    plt.title(f"Ventas por producto - {nombre.upper()}", fontsize=14)
    plt.xlabel("Cantidad vendida")
    plt.ylabel("Producto")
    plt.tight_layout()
    plt.grid(axis='x', linestyle='--', alpha=0.5)
    plt.show()
```

# 5. Envío promedio por tienda

```python
# Iterar sobre cada tienda
tiendas = {
    'tienda': tienda,
    'tienda2': tienda2,
    'tienda3': tienda3,
    'tienda4': tienda4
}
for nombre, df in tiendas.items():
    costo_envio_prom = df['Costo de envío'].value_counts()
    costo_envio_promedio = (costo_envio_prom * costo_envio_prom.index).sum() / costo_envio_prom.sum()
    print(f"Costo de envío promedio de {nombre}: {costo_envio_promedio:.2f}")

```
